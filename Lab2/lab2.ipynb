# Exercitiul 1

```python
# Punctul a)
def extrage_bila():
    # declaram urna
    urna = ["red"] * 3 + ["blue"] * 4 + ["black"] * 2
    # dam cu zarul
    zar = np.random.randint(1, 7)

    # adaugam o bila in functie de numarul zarului
    if zar == 2 or zar == 3 or zar == 5:
        urna += ["black"]
    elif zar == 6:
        urna += ["red"]
    else:
        urna += ["blue"]

    # alegem o bila din urna
    index = np.random.randint(0, len(urna))
    return urna[index]


# Punctul b)
# repetam experimentul de mai sus de 1000 de ori
count_red = 0
for index in range(0, 1000):
    if extrage_bila() == "red":
        count_red+=1
print(f"Probabilitatea de a alege o bila rosie este: {count_red/1000:.3f}")
```

D = extragem o bila rosie

A = numarul zarului este prim
P(A) = 3/6 {2, 3, 5}
Probabilitatea sa extragem o bila rosie devine P(D|A) = 3/10

B = numarul zarului este 6
P(B) = 1/6 {6}
Probabilitatea sa extragem o bila rosie devine P(D|B) = 4/10

C = numarul zarului este {1, 4}
P(C) = 2/6
Probabilitatea sa extragem o bila rosie devine P(D|C) = 3/10

P(D) = P(D|A) * P(A) + P(D|B) * P(B) + P(D|C) * P(C)
     = 3/6 * 3/10 + 1/6 * 4/10 + 2/6 * 3/10
     = 1/10 * (3/2 + 2/3 + 1)
     = 1/10 * (9 + 4 + 6) * 1/6
     = 19/60
     = 0.316

Observam ca probabilitatea de la punctul b este foarte apropiata de cea teoretica obtinuta la punctul c

```python
# Punctul 1
# Simulam 1000 de valori dintr-o distributie Poisson cu lambda = 1
poisson_random_1 = np.random.poisson(lam=1, size=1000)

# Simulam 1000 de valori dintr-o distributie Poisson cu lambda = 2
poisson_random_2 = np.random.poisson(lam=2, size=1000)

# Simulam 1000 de valori dintr-o distributie Poisson cu lambda = 5
poisson_random_5 = np.random.poisson(lam=5, size=1000)

# Simulam 1000 de valori dintr-o distributie Poisson cu lambda = 10
poisson_random_10 = np.random.poisson(lam=10, size=1000)


# Punctul 2
lam_posibil = [1, 2, 5, 10]

# Simulam 1000 de valori dintr-o distributie Poisson cu lambda random
poisson_random = []
for index in range(0, 1000):
    #alegem un lambda random pentru fiecare valoare
    lam = random.choice(lam_posibil)
    value = np.random.poisson(lam=lam, size=1)
    poisson_random.append(value[0])


# Punctul 2 a)
# Afisam histograma distributiei Poisson
plt.hist(poisson_random_1, bins=15, color='orange', edgecolor='black', alpha=0.7)
plt.title('Distribuția Poisson (lambda = 1)')
plt.xlabel('Valori')
plt.ylabel('Frecventa')
plt.show()
```

```python
plt.hist(poisson_random_2, bins=15, color='orange', edgecolor='black', alpha=0.7)
plt.title('Distribuția Poisson (lambda = 2)')
plt.xlabel('Valori')
plt.ylabel('Frecventa')
plt.show()
```

```python
plt.hist(poisson_random_5, bins=15, color='orange', edgecolor='black', alpha=0.7)
plt.title('Distribuția Poisson (lambda = 5)')
plt.xlabel('Valori')
plt.ylabel('Frecventa')
plt.show()
```

```python
plt.hist(poisson_random_10, bins=15, color='orange', edgecolor='black', alpha=0.7)
plt.title('Distribuția Poisson (lambda = 10)')
plt.xlabel('Valori')
plt.ylabel('Frecventa')
plt.show()
```

```python
plt.hist(poisson_random, bins=15, color='orange', edgecolor='black', alpha=0.7)
plt.title('Distribuția Poisson (lambda random)')
plt.xlabel('Valori')
plt.ylabel('Frecventa')
plt.show()
```
