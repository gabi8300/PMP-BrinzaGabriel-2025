```python
import pymc as pm
import arviz as az
import matplotlib.pyplot as plt

Y = [0, 5, 10]
Teta = [0.2, 0.5]

fig, axes = plt.subplots(len(Y), len(Teta), figsize=(10, 10))

for i, y in enumerate(Y):
    for j, teta in enumerate(Teta):
        with pm.Model() as model:
            n = pm.Poisson("n", 10)
            Y_obs = pm.Binomial("Y_obs", n=n, p=teta, observed=y)
            trace = pm.sample(
                draws=2000,
                tune=2000,
                chains=2,
                step=pm.Metropolis(),
                progressbar=False
            )

        az.plot_posterior(trace, var_names=["n"], ax=axes[i, j])
        axes[i,j].set_title(f"Y={y}, ={teta}")

plt.tight_layout()
plt.show()
```
